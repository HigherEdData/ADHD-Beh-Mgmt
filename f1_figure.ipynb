```stata
ssc install dubois
```

```stata
use data/Cleaned_Data.dta, clear
```

```stata
qui reg dummy_contact i.child_adhd, ro

qui margins i.child_adhd, atmeans
```

```stata
marginsplot, scheme(dubois) ///
title("Figure 1: School Contact by Diagnosis" " ") ///
    ytitle(students with ADHD)
```

```stata
graph save School_Cnt_Diag, replace
```

```stata
graph export data/School_Cnt_Diag.pdf, replace
```
